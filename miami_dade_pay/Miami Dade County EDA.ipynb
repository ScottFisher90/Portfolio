```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Employee_Pay_Information.csv')
df.head()
```

<h2>Salary info for each department</h2>

```python
df['FullName'] = df['FirstName'] + ' ' + df['LastName']
#df.info()
df1 = df[['FullName', 'Title', 'DeptName', 'AnnualSalary']]
df2 = df1.groupby(['DeptName'])['AnnualSalary'].agg(['mean', 'median', 'min', 'max', 'count'])
df2
```

```python
filter = df1['AnnualSalary'] > 100000
high_salary_df = df1[filter]
high_salary_df.sort_values(by=['AnnualSalary'], ascending=False)
```

```python
total_salary = df.AnnualSalary.sum()


df3 = pd.DataFrame(df1.groupby(['DeptName'])['AnnualSalary'].sum())
df3['percentage_of_total_wages'] = df3['AnnualSalary']/total_salary *100
df3.sort_values(by=['percentage_of_total_wages']) 
df4 = df3.merge(df2, on='DeptName')
df4.drop(['mean', 'median', 'min', 'max'], axis=1, inplace=True)
total_employees = df4['count'].sum()
df4['percent_of_total_employees'] = df4['count']/total_employees * 100
df4['wage_inflation_metric'] = df4['percentage_of_total_wages']/df4['percent_of_total_employees']
df4.sort_values(by=['percentage_of_total_wages', 'count'])
```

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.scatterplot(
    data=df4,
    x='percent_of_total_employees',
    y='percentage_of_total_wages'
)
```

Note the linear relationship. The wage_inflation_metric is a number depicting how much the department differs from the linear relationship. If perfectly linear, each would be equal to 1, meaning that the total amount of funding the department is given scales equally with employee numbers throughout the county. Department of interest: County Attorney's Office, Inspector General, Mayor Office, Transportation Trust, Ethics Commision.
